```python
bus_locations = pd.read_csv('bus_locations_1238_pm_8.2.2019.csv')
```

## Joins

```python
M11_stops['stop_id'] = M11_stops['stop_id'].astype(str)
bus_stops['stop_id'] = bus_stops['stop_id'].astype(str)
```

```python
m11_stop_locations = M11_stops.merge(bus_stops, left_on = 'stop_id', right_on='stop_id')
```

# Folium Time

```python
m11_stop_locations['stop_lon'] =m11_stop_locations['stop_lon'].astype(float)
m11_stop_locations['stop_lat'] =m11_stop_locations['stop_lat'].astype(float)
```

```python
#Create a list of points travelled by the vehicle
geo_list = list(zip(m11_stop_locations['stop_lat'],m11_stop_locations['stop_lon']))
```

```python
#List of times stoped by the vehicle
trip_list = list(range(len(m11_stop_locations)))
```

```python
import base64
import io

encoded = base64.b64encode(open('bus.png', 'rb').read())
decoded = base64.b64decode(encoded)
icon_url = io.BytesIO(decoded)
bus_image = folium.features.CustomIcon(icon_url, icon_size=(100,100))
```

```python
bus_image
```

```python
#Visualize the basemap
m = folium.Map(location=[40.783, -73.95], zoom_start = 11)

#Create the color line of the bus route
one = folium.ColorLine(geo_list,
                 colors = trip_list,
                 colormap = ['blue', 'red'],
                 weight = 25,
                 opacity = 0.8,
                 legend_name = 'Bus Route Start to Finish').add_to(m)


# add a name to the layers
one.layer_name = 'Bus Route'
fg = folium.FeatureGroup(name='Bus Stop Locations')
bg = folium.FeatureGroup(name='Bus Locations')

#For loop to add stop points to fg
for point in range(0, len(geo_list)):
    fg.add_child(folium.Marker(geo_list[point], popup=m11_stop_locations['stop_name'][point]))
    
#For loop to add bus location points to fg
for point in range(0, len(bus_locations_list)):
    bg.add_child(folium.Marker(bus_locations_list[point], 
                               icon=folium.Icon(color='green'), 
                               popup='Bus #'+bus_locations['Bus Number'][point]))

    
# Overlay the images
m.add_children(one)
m.add_children(fg)
m.add_children(bg)
    
# add the layer control
folium.LayerControl().add_to(m)

m
```
